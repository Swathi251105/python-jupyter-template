### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Line Plot
x = np.linspace(0, 10, 100)
y = x ** 2
axes[0, 0].plot(x, y, label='y = x^2', color='b', linestyle='-', linewidth=2)
axes[0, 0].set_title('Line Plot')
axes[0, 0].set_xlabel('X Values')
axes[0, 0].set_ylabel('Y Values')
axes[0, 0].grid(True, linestyle='--', alpha=0.6)
axes[0, 0].legend()

# Bar Plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']
axes[0, 1].bar(categories, values, color=colors)
axes[0, 1].set_title('Bar Plot')
axes[0, 1].set_xlabel('Categories')
axes[0, 1].set_ylabel('Values')

# Scatter Plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axes[1, 0].scatter(x_scatter, y_scatter, color='red')
for i, txt in enumerate(zip(x_scatter, y_scatter)):
    axes[1, 0].annotate(txt, (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(5,5), ha='right')
axes[1, 0].set_title('Scatter Plot')
axes[1, 0].set_xlabel('X Values')
axes[1, 0].set_ylabel('Y Values')

# Pie Chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'red', 'green', 'orange'])
axes[1, 1].set_title('Pie Chart')

plt.tight_layout()
plt.show()

# Histogram
plt.figure(figsize=(8, 5))
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]
plt.hist(data, bins=5, color='teal', edgecolor='black')

# Customizations
plt.title('Histogram of Data')
plt.xlabel('Data Values')
plt.ylabel('Frequency')

# Show the histogram
plt.show()
```
