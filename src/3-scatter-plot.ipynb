### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
import numpy as np
import matplotlib.pyplot as plt

def validate_data(x, y):
    if not isinstance(x, (list, np.ndarray)) or not isinstance(y, (list, np.ndarray)):
        raise ValueError("Input data must be a list or numpy array.")
    if len(x) != len(y):
        raise ValueError("x and y must have the same length.")

def plot_line(x, y):
    validate_data(x, y)
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label='y = x^2', color='b', linestyle='-', linewidth=2)
    plt.title('Line Plot of y = x^2')
    plt.xlabel('X Values')
    plt.ylabel('Y Values')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

def plot_bar(categories, values, colors):
    if not (isinstance(categories, list) and isinstance(values, list) and isinstance(colors, list)):
        raise ValueError("Categories, values, and colors must be lists.")
    if len(categories) != len(values) or len(categories) != len(colors):
        raise ValueError("Categories, values, and colors must have the same length.")
    plt.figure(figsize=(8, 5))
    plt.bar(categories, values, color=colors)
    plt.title('Bar Plot of Categories')
    plt.xlabel('Categories')
    plt.ylabel('Values')
    plt.show()

def plot_scatter(x, y):
    validate_data(x, y)
    plt.figure(figsize=(8, 5))
    plt.scatter(x, y, color='red')
    for i, txt in enumerate(zip(x, y)):
        plt.annotate(txt, (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='right')
    plt.title('Scatter Plot with Annotations')
    plt.xlabel('X Values')
    plt.ylabel('Y Values')
    plt.show()

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)
y = x ** 2  # Square of x
plot_line(x, y)

# Create a bar plot
tags = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']
plot_bar(tags, values, colors)

# Create a scatter plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
plot_scatter(x_scatter, y_scatter)
```
