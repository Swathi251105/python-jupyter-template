### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)
y = x ** 2  # Square of x

# Create the line plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='y = x^2', color='b', linestyle='-', linewidth=2)

# Customizations
plt.title('Line Plot of y = x^2')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Show the line plot
plt.show()

# Create a bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']

plt.figure(figsize=(8, 5))
plt.bar(categories, values, color=colors)

# Customizations
plt.title('Bar Plot of Categories')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show the bar plot
plt.show()

# Create a scatter plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

plt.figure(figsize=(8, 5))
plt.scatter(x_scatter, y_scatter, color='red')

# Annotate each point
for i, txt in enumerate(zip(x_scatter, y_scatter)):
    plt.annotate(txt, (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(5,5), ha='right')

# Customizations
plt.title('Scatter Plot with Annotations')
plt.xlabel('X Values')
plt.ylabel('Y Values')

# Show the scatter plot
plt.show()

# Create a pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

plt.figure(figsize=(8, 5))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'red', 'green', 'orange'])

# Customizations
plt.title('Pie Chart of Programming Languages')

# Show the pie chart
plt.show()
```
